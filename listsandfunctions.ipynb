```python
#9.2
#for loop to check if number is divisible by 2
#if not, yield the number
def get_odds():
    for num in range(10):
        if num % 2 != 0:
            yield num

#start a count at 0 and loop until the count is 3
#print the 3rd odd number
count = 0
for value in get_odds():
    count += 1
    if count == 3:
        print(value)
        break
```
